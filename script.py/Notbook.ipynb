```python

import pandas as pd

# 1 - Read and parse CSV files into pandas DataFrames with dates parsed as datetime objects.
visits = pd.read_csv('../raw_files/visits.csv', parse_dates=[1])
cart = pd.read_csv('../raw_files/cart.csv', parse_dates=[1])
checkout = pd.read_csv('../raw_files/checkout.csv', parse_dates=[1])
purchase = pd.read_csv('../raw_files/purchase.csv', parse_dates=[1])

# 2 - Merge the 'visits' and 'cart' DataFrames using a left join and store in visits_cart.
visits_cart = visits.merge(cart, how='left')

# 3 - Calculate the total number of visits after merging.
total_visits = len(visits_cart)
print(total_visits)

# 4 - Calculate the number of visits where the cart_time is null.
null_cart_times = len(visits_cart[visits_cart.cart_time.isnull()])
print(null_cart_times)

# 5 - Calculate the percentage of visitors who visited but did not place anything in their cart.
visited_not_cart = float(null_cart_times) / float(total_visits)
print(f'A total of {visited_not_cart}% of the visitors ended up not placing anything in their cart')

# 6 - Merge the 'cart' and 'checkout' DataFrames using a left join and store in cart_checkout.
cart_checkout = cart.merge(checkout, how='left')
print(cart_checkout)

# 7 - Merge all relevant DataFrames ('visits_cart', 'cart_checkout', 'purchase') into a single DataFrame 'all_data'.
all_data = visits_cart.merge(cart_checkout, how='left').merge(purchase, how='left')
print(all_data.head(5))



```
